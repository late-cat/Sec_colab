```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv("Churn_Modelling.csv")
dataset.head()
```

## dividing the data set into dependent and independent features

```python
X=dataset.iloc[:,3:13]
Y=dataset.iloc[:,13]
```

```python
X.head()
```

```python
Y
```

##Feature engeneering

```python
geography=pd.get_dummies(X['Geography'],drop_first=True)
gender=pd.get_dummies(X['Gender'],drop_first=True)
```

```python
##concatanate into the data frame
X=X.drop(['Geography','Gender'],axis=1)
```

```python
X=pd.concat([X,geography,gender],axis=1)
```

```python
##splitting dataset into test train
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
##fetaure scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
X_train.shape
```

```python
##creat The ANN
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LeakyReLU,PReLU,ELU,ReLU
from tensorflow.keras.layers import Dropout
```

```python
from tensorflow.keras.layers import Input
classifier = Sequential()
classifier.add(Input(shape=(11,)))
```

```python
##adding the inp layer
classifier.add(Dense(units=11, activation='relu'))
```

```python
##adding the first hidden layer
classifier.add(Dense(units=7, activation='relu'))
```

```python
##adding 2nd HL
classifier.add(Dense(units=6, activation='relu'))
```

```python
## adding O/P layer
classifier.add(Dense(1, activation='sigmoid'))
```

```python
##compile the model
classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
import tensorflow
opt=tensorflow.keras.optimizers.Adam(learning_rate=0.01)
```

```python
##early stopping
import tensorflow as tf
from tensorflow import keras
early_stopping=keras.callbacks.EarlyStopping(
    monitor="val_loss",
    min_delta=0.0001,
    patience=10,
    verbose=1,
    mode="auto",
    baseline=None,
    restore_best_weights=False,
    start_from_epoch=0,
)
```

```python
# Compiling the model before training to ensure valid state
classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Training the ANN with Early Stopping callback
model_history = classifier.fit(
    X_train,
    Y_train.flatten(),
    validation_split=0.33,
    batch_size=10,
    epochs=1000,
    callbacks=[early_stopping]
)
```

```python
model_history.history.keys()
```

```python
# summarize history for accuracy
import matplotlib.pyplot as plt

plt.plot(model_history.history['accuracy'])
plt.plot(model_history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
# summarize history for loss
plt.plot(model_history.history['loss'])
plt.plot(model_history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
##predict the test set results
Y_predict=classifier.predict(X_test)
Y_predict=(Y_predict>=0.5)
```

```python
##calc the accuracy
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,Y_predict)
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(Y_test, Y_predict)
```
